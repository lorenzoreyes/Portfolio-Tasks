# Use fisher library for tech analysis
# & backtest indicators deployment

```python
import pandas as pd
import numpy as np 
import yfinance as yahoo

import fisher
```

# Use Coca-Cola Stock for the examples
# Fisher Simple-Moving-Average

```python
cokeSMA = fisher.Strategy.SimpleMovingAverage(symbol='KO', num_days=100)
cokeSMA
# signal is a conditional if SMA-period > StockPrice buy, else sell
# positions tracks signal differences
```

```python
cokeSMA.iloc[:,:2].plot(figsize=(12,6),grid=True, lw=3.0)
```

# Fisher-Momentum

```python
cokeMOM = fisher.Strategy.Momentum(symbol='KO',num_days=50)
cokeMOM
```

```python
cokeMOM['Momentum'].plot(figsize=(12,6),grid=True,legend=True, lw=3.0)
```

# Fisher-Bollinger-Bands

```python
cokeBB = fisher.Strategy.BollingerBands(symbol='KO',num_days=50,stdev_factor=2)
cokeBB
```

```python
cokeBB.plot(figsize=(12,6),grid=True,lw=3.0)
```

# Fisher-Standard-Deviation

```python
cokeStDev = fisher.Strategy.StandardDeviation(symbol='KO',num_days=50)
cokeStDev
```

```python
cokeStDev['StDev50'].plot(figsize=(12,6),grid=True,legend=True,lw=3.0)
```

# Fisher-Absolute-Price-Oscillator

```python
cokeAPO = fisher.Strategy.AbsolutePriceOscillator(symbol='KO',short=20,long=100)
cokeAPO
```

```python
cokeAPO.iloc[:,1:].plot(figsize=(12,6),grid=True,lw=3.0)
```

# Fisher-Relative-Strength-Index-(RSI)

```python
cokeRSI = fisher.Strategy.RelativeStrenghtIndex(symbol='KO',time_period=50)
cokeRSI
```

```python
cokeRSI['KO RSI'].plot(figsize=(12,6),grid=True,lw=3.0,legend=True)
```

# Fisher-Moving-Average-Convergence-Divergence-(MACD)

```python
cokeMACD = fisher.Strategy.MovingAverageConvergenceDivergence(symbol='KO',short=20,long=100)
cokeMACD
```

```python
cokeMACD['MACD KO'].plot(figsize=(12,6),grid=True,legend=True,lw=3.0)
```
